# Summary Data

```python
path = "/home/ada/hub/FifaAnalysis/data/"

dfs = []

for i in os.listdir(path):
    if "Summary.csv" not in i:
        df = pd.read_csv(path + i)
        df["year"] = int(i.split("-")[-1].split(".")[0])
        dfs.append(df)
        
df = pd.concat(dfs)
print(df.shape)
df.sample(5)
```

```python
summ = pd.read_csv(path + "FIFA - World Cup Summary.csv")
summ.sample()
```

- Average comparison: yrly,
